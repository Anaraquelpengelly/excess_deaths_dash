```python
#reset years
years=list(range(2014, 2021))
new_df_list=[]
# rename the row for each data frame for each year
## reset index to 'Week ended'
for i in range(len(years)):
    t=df_list[i].set_index('Week number')
    t.rename(index={'Total deaths, all ages': 'total_deaths_'+str(years[i])}, inplace=True)
    new_df_list.append(t)

#remove rows above and below the wanted rows


#make df of those rows
```

```python
list_tot=[new_df_list[i].loc['total_deaths_'+str(years[i])] for i in range(len(new_df_list))]
list_tot
df_tot=pd.DataFrame(list_tot)
df_tot=df_tot.drop(['Unnamed: 1'], axis=1)
df_tot
```

```python
df_tot_t=df_tot.transpose()
df_tot_t.head()
```

```python
list(df_tot_t.columns[:-1])
```

```python
# Now we calculate the average for 2014-2019
df_tot_t['avg_weekly_deaths_14_19']=df_tot_t[list(df_tot_t.columns[:-1])].mean(axis=1)
df_tot_t.head()
```

```python
df_tot_t['week_number']=range(1, 54)
df_tot_t
```

```python
#difference between the average 2014-2019 and 2020
df_tot_t['diff_avg_14_19_to_20']= df_tot_t['total_deaths_2020']-df_tot_t['avg_weekly_deaths_14_19']
df_tot_t
```

```python
total_excess_deaths=np.sum(df_tot_t['diff_avg_14_19_to_20'])
total_excess_deaths
```

## Compare the number of deaths in 2020 until week 43 and the average number of deaths from 2014-2020 to week 43

```python
test=df_tot_t[df_tot_t['week_number'].isin(list(range(1, 44)))]
```

```python
avg_tot_deaths_14_19_w43=np.sum(test['avg_weekly_deaths_14_19'])
tot_deaths_20_w43=np.sum(df_tot_t['total_deaths_2020'])
```

```python
print(f'the total number of deaths in 2020 is so far: {tot_deaths_20_w43}, \nthe the average yearly deaths from 2014 to 2019 up to week 43 is: {avg_tot_deaths_14_19_w43}. \nThe difference is: {round(tot_deaths_20_w43-avg_tot_deaths_14_19_w43, 2)}')
```

```python
print(f'The percentage of excess deaths in 2020 is: {round((tot_deaths_20_w43-avg_tot_deaths_14_19_w43)/tot_deaths_20_w43*100, 2)}%')
```

# Now plot both the average deaths in 2014-2019 and the deaths in 2020 in the same plot!

```python
##use this : 

df1 = df_tot_t[['total_deaths_2020', 'week_number']]

df2 = df_tot_t[['avg_weekly_deaths_14_19', 'week_number']]

chart1 = alt.Chart(df1).mark_line(color='red').encode(
    x=alt.X('week_number', title='Week number'), 
    y=alt.Y('total_deaths_2020', title='Deaths')
)

chart2 = alt.Chart(df2).mark_line().encode(
    x='week_number', 
    y='avg_weekly_deaths_14_19')

chart1 + chart2
```

```python
# Now lets try this:
df=df_tot_t[['week_number', 'total_deaths_2020', 'avg_weekly_deaths_14_19']]
alt.Chart(df).mark_line().transform_fold(
    fold=['total_deaths_2020', 'avg_weekly_deaths_14_19'], 
    as_=['variable', 'value']
).encode(
    x=alt.X('week_number', title='Week number'),
    y=alt.Y('value:Q', title='Deaths'),
    color=alt.Color('variable:N', legend=alt.Legend(title=None))
)
```
